```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the created plots

# Read in the data (.csv)

df = pd.read_csv(r'C:\Users\janni\Downloads\movies.csv') #df=dataframe
```

```python
# Display some data
df.head()
```

```python
# Drop the missing data
df = df.dropna()
```

```python
# Check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types of columns
df.dtypes
```

```python
# Adjust column datatypes
df['budget'] = df['budget'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# Year and year released should match -> adjust by extracting 4 digit number pattern
df['yearcorrect'] =  df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df = df.sort_values(by = ['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', 20)
```

```python
# Check for duplicate rows
df2 = df[df.duplicated()]
print (df2)
# Drop duplicates if needed
df.drop_duplicates()
```

```python
# Hypothesis: Budget has high correlation to gross revenue 
# Comapny has high correlation to gross revenue
```

```python
# Scatter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Revenue')
plt.xlabel('Film Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Create correlation heatmap
correlation_matrix = df.corr(method= 'pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
df_new = df[["yearcorrect","budget","votes","gross","runtime"]]
df_new.head()

correlation_matrix = df_new.corr(method= 'pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
#Include company information (string) in correlation matrix as numeric value
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

```

```python
correlation_matrix = df_numerized.corr(method= 'pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
corr_mat = df_numerized.corr()
corr_pairs = corr_mat.unstack()
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs > 0.5)]
pd.set_option('display.max_rows', None)
high_corr

# Correlation between gross revnue and buget and votes is high
```
